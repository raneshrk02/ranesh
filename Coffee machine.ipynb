```python
print("PyCoffee Machine!!!!")
employee = ["ranesh","arun","keerthivasan","rahul","eshwar","harshit","mahesh"]
rep = {
    "water" : 500,
    "coffee" : 200,
    "milk" : 400,
    "money" : 0
}
should_continue = True
id_continue = True
while should_continue == True:
  
  req = str(input("What you want today(coffee/report/refill) ")).lower()

  if req == "coffee":
    coff = str(input("What coffee do you want(espresso/latte/cappuchino) ")).lower()
    
    if coff == "espresso":
      if rep["water"] < 50 or rep["coffee"] < 18:
        print("Sorry, not enough items available to make your coffee :(")
      else:
        print("That would be ₹20")
        tens = int(input("How many tens? "))
        hundreds = int(input("How many hundreds? "))
        cost = hundreds*100 + tens*10
        if cost > 20:
          print(f"Please collect back ₹{cost-20}. And dont forget your coffee!!!")
          rep["water"] -= 50
          rep["coffee"] -= 18
          rep['money'] += 20 
        elif cost < 20:
          print(f"Sorry, insufficient money deposited. You should have deposited {20-cost} more. Money refunded.")
        elif cost == 20:
          print("Please collect your coffee!! Have a nice day!!")
          rep["water"] -= 50
          rep["coffee"] -= 18
          rep['money'] += 20 

    elif coff == "latte":
      if rep["water"] < 200 or rep["coffee"] < 24 or rep["milk"] < 150:
        print("Sorry, not enough items available to make your coffee :(")
      else:
        print("That would be ₹40")
        tens = int(input("How many tens? "))
        hundreds = int(input("How many hundreds? "))
        cost = hundreds*100 + tens*10
        if cost > 40:
          print(f"Please collect back ₹{cost-40}. And dont forget your coffee!!!")
          rep["water"] -= 200
          rep["coffee"] -= 24
          rep["milk"] -= 150
          rep['money'] += 40
        elif cost < 40:
          print(f"Sorry, insufficient money deposited. You should have deposited {40-cost} more. Money refunded.")
        elif cost == 40:
          print("Please collect your coffee!! Have a nice day!!")
          rep["water"] -= 200
          rep["coffee"] -= 24
          rep["milk"] -= 150
          rep['money'] += 40
    
    elif coff == "cappuchino":
      if rep["water"] < 250 or rep["coffee"] < 24 or rep["milk"] < 100:
        print("Sorry, not enough items available to make your coffee :(")
      else:
        print("That would be ₹60")
        tens = int(input("How many tens? "))
        hundreds = int(input("How many hundreds? "))
        cost = hundreds*100 + tens*10
        if cost > 60:
          print(f"Please collect back ₹{cost-60}. And dont forget your coffee!!!")
          rep["water"] -= 250
          rep["coffee"] -= 24
          rep["milk"] -= 100
          rep['money'] += 60
        elif cost < 60:
          print(f"Sorry, insufficient money deposited. You should have deposited {60-cost} more. Money refunded.")
        elif cost == 60:
          print("Please collect your coffee!! Have a nice day!!")
          rep["water"] -= 250
          rep["coffee"] -= 24
          rep["milk"] -= 100
          rep['money'] += 60

  if req == "report":
    while id_continue == True:
      id = str(input("Enter your employee code: "))
      if id not in employee:
        print("Invalid credentials")
      else:
        print(f"Water : {rep['water']} ml")
        print(f"Coffee : {rep['coffee']} g")
        print(f"Milk : {rep['milk']} ml")
        print(f"Money : ₹{rep['money']}")
        id_continue == False
        break

  if req == "refill":
    while id_continue == True:
      id = str(input("Enter your employee code: "))
      if id not in employee:
        print("Invalid credentials")
      else:
        rep['water'] = 500
        rep['coffee'] = 200
        rep['milk'] = 400
        print(f"Water : {rep['water']} ml")
        print(f"Coffee : {rep['coffee']} g")
        print(f"Milk : {rep['milk']} ml")
        print(f"Money : ₹{rep['money']}")
        id_continue == False
        break

  if req == "nothing":
    while id_continue == True:
      id = str(input("Enter your employee code: "))
      if id not in employee:
        print("Invalid credentials")
      else:
        print(f"The day's earnings: ₹{rep['money']}")
        id_continue == False
        break
    should_continue == False
    break
```
