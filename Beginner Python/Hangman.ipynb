```python
import random
stages = ['''
 +---+
 |   |
 O   |
/|\  |
/ \  |
     |
 ===== 
''', '''
 +---+
 |   |
 O   |
/|\  |
  \  |
     |
 ===== 
''', '''
 +---+
 |   |
 O   |
/|\  |
     |
     |
 ===== 
''','''
 +---+
 |   |
 O   |
/|   |
     |
     |
 ===== 
''','''
 +---+
 |   |
 O   |
 |   |
     |
     |
 ===== 
''', '''
 +---+
 |   |
 O   |
     |
     |
     |
 ===== 
''', '''
 +---+
 |   |
     |
     |
     |
     |
 ===== 
''']

l1 = ["apple", "sad", "date"]
chosen_word = random.choice(l1)
lives = 6
display = []
for i in range(len(chosen_word)):
  display += "_"
print(f"{' '.join(display)}")
end = False
while not end:
  n = input("Guess a letter: ").lower()
  for i in range(len(chosen_word)):
    x = chosen_word[i]
    if x == n:
      display[i] = x
  
  if n not in chosen_word:
    lives -= 1
    print(stages[lives])  
    if lives == 0:
      end = True
      print("You lose")
  
  print(f"{' '.join(display)}")
  
  if "_" not in display:
    end = True
    print("You win")
```
